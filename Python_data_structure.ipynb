1> Discuss string slicing and provide example?

Answre:- String slicing is the process of extracting a specific part of string with the help of indexes of the string. String slicing is an important aspect of string manipulation.
         String slicing can be performed by writing the string identifier and then mentioning the index number from where the slicing should begin and where to finish.
         
         The syntax for string slicing is:-
             
             string[start:stop:step]
             
                     Here, start represents the index value from where the slicing to begin.
                           stop represent the index value till where the slicing will continue.
                           step represents the number of jump values/ the index value to be skipped during the process by caluclation.
                           
             Example:-
                     
myname="Sritam Dash"
myname[0:7:2]

5> Describe the key features of sets and provide example of their use?

Answer:- 

Uniqueness and No Duplicates:
               
     Sets contain unique elements. Each element appears only once within a set.
                
     If you try to add a duplicate element, it won’t be included.
                
Unordered:

    Unlike lists, sets are unordered. This means that the order of elements in a set is not guaranteed.
    
    Sets don’t support indexing or slicing.

Immutable Elements:
        
    Set elements must be of immutable types (e.g., numbers, strings, tuples).
    
    Mutable types like lists or dictionaries cannot be elements of a set.

Optimized for Membership Testing:
    
    Sets have an efficient method for checking whether an element is present.
    
    This makes them useful for tasks like filtering unique values or testing membership.
    
Examples:-

    Creating Sets:

    We create sets by enclosing elements in curly braces {} and separating them with commas.
    Example:

    # Creating sets
    student_ids = {112, 114, 116, 118, 115}
    vowel_letters = {'a', 'e', 'i', 'o', 'u'}
    mixed_set = {'Hello', 101, -2, 'Bye'}
    
    Adding Items to a Set:
    
    To add an item to a set, use the add() method.
    Example:
    
    numbers = {21, 34, 54, 12}
    numbers.add(32)  # Adds 32 to the set

7> Describe how to add, modify, delete data from dictionary with example?

Answer:-
        Adding Items to a Dictionary:
            
            Using Assignment:
            
            To add a new key-value pair to an existing dictionary, simply assign a value to a new key:

            my_dict = {'name': 'Alice', 'age': 30}
            my_dict['city'] = 'Wonderland'  # Adding a new key-value pair
            
            Using the update() Method:
            
            The update() method allows you to merge another dictionary (or an iterable of key-value pairs) into an existing dictionary:

            current_library = {'Python Crash Course': 30, 'Data Science for Beginners': 25}
            new_books = {'Advanced Python Programming': 15, 'Machine Learning Essentials': 20}
            current_library.update(new_books)  # Adding new books
            
        Modifying Items in a Dictionary:

            Direct Assignment:
            To modify the value associated with an existing key, simply assign a new value to that key:

            my_dict = {'name': 'Alice', 'age': 30}
            my_dict['age'] = 31  # Modifying the value for the 'age' key
            
        Deleting Items from a Dictionary:
            
            Using del Keyword:
            To remove a key-value pair from a dictionary, use the del keyword:
            
            my_dict = {'name': 'Alice', 'age': 30}
            del my_dict['age']  # Deleting the 'age' key
            
            Using pop() Method:
            
            The pop() method removes a key and returns its value. You can also provide a default value if the key doesn’t exist:

            my_dict = {'name': 'Alice', 'age': 30}
            removed_age = my_dict.pop('age')  # Removing and getting the value for 'age'

8> Discuss the importance of dictionary key being immutable and provide example?

Answer:- 

Hashability and Dictionary Keys:
                
            . Immutable objects can be used as keys in dictionaries because their hash value remains constant throughout their lifetime.
            . When you create a dictionary with immutable keys, Python computes a hash value for each key. This hash value is used to efficiently look up values associated with 
              those keys.
            . Since the hash value doesn’t change, the key-value mapping remains consistent.
            . This property ensures that dictionaries can quickly retrieve values based on their keys without recalculating hashes
            
            example:-
                # Using strings (immutable) as dictionary keys
                user_info = {
                    'username': 'alice',
                    'age': 30,
                    'email': 'alice@example.com'
                }

Memory Efficiency:
        
            . Immutable keys contribute to memory efficiency.
            . When you use the same immutable key multiple times in different dictionaries, Python can optimize memory usage by reusing the same key object.
            . This reduces memory overhead compared to mutable keys, which would require separate memory allocations for each occurrence.
            
            example:-
            
                # Reusing the same string as keys in different dictionaries
                user1 = {'username': 'alice', 'age': 30}
                user2 = {'username': 'alice', 'email': 'alice@example.com'}

Predictability and Debugging:

            . Immutability ensures that a given key’s value will not change unexpectedly.
            . When you work with dictionaries containing immutable keys, you can reason about their behavior more confidently.
            . Debugging becomes easier because you don’t need to track potential changes to keys
            
            example:-
            
            # Assume 'user_info' is a dictionary with immutable keys
            if 'username' in user_info:
                print(f"Username: {user_info['username']}")
